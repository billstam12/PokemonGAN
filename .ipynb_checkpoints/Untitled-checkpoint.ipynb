# Generative Adverserial Network to create new Pokemon sprites
## Created using pyTorch

In this notebook I will be creating a GAN which will use a dataset containing over 14000 pokemon sprites and after training will try to create totally new ones.

In the cells below I will describe what I did, while I will explain what is a GAN and what is it's goal.

```python
import torch
from torch import nn, optim
from torch.autograd.variable import Variable
from torchvision import transforms, datasets
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Generate the data from the dataset in my folder

def pokemon_data():
    compose = transforms.Compose(
        [transforms.Resize(64),
         transforms.CenterCrop(64),
         transforms.ToTensor(),
         transforms.Normalize((.5, .5, .5), (.5, .5, .5))
        ])
    out_dir = 'mgan_dataset'
    return datasets.ImageFolder(root=out_dir,transform=compose)

data = pokemon_data()
poke_loader = torch.utils.data.DataLoader(data, batch_size=64, shuffle=True)
# Num batches
num_batches = len(poke_loader)
```

```python
#print a single image

for img in poke_loader:
    single_image = img[0][0]
    single_image= (single_image*0.5)+0.5
    single_image = single_image.clamp(0,1)
    single_image = single_image.numpy()
    
    # move the dimensions around to get them right
    single_image = np.transpose(single_image, (1, 2, 0))
   
    # plot image
    print('image size:' ,single_image.shape)
    plt.imshow(single_image)
    plt.axis('off')
    plt.show()
    break
```

## Generator

Here we create the **generator** neural network, which will try to create new image data like the above. It's goal will be to imitate the input data as good as it can.

Mathematically a neural network *G(z,θ1)* will be used to map noise data *z* to the real input data *x*.
θ1 represents the weights of the parameters of the network. It's goal can be represented as it trying to maximize the probability of a **fake** image being classified as real. The one that chooses which images are real or not is another Neural Network called the **Discriminator**. The loss function we want to maximize is thus ***D(G(z))***, and as with most NNs we will take the log of this loss functions resulting in *max(log(D(G(z))))*

## Discriminator

After creating the generator model we will create the discriminator, both using pyTorch. This networks goal will be to increase the probability of correctly classifying real data to the original dataset and flagging forged data of the generator as false. This means that it's existence depends on making the generator's life as difficult as possible, while the latter tries to make the discriminators life as hard as possible too.
Taking this things into account, we can understand that the two models will be playing a game (that's why the **adverserial** exists in the name) and will be trying to one up one another. The discriminator will want to minimize the above loss function, in contrast to the generator, making the game a minimax like below:



The model's final goal will be to reach a point where no improvement from both sides can happen and thus the model will converge and it's training will be complete.


Below we define the Discriminator model, which is a **6 hidden layer** Network with a *LeakyRelu* function for normalization and a 0.3 Dropout to prevent overfitting.

```python
class Discriminator(torch.nn.Module):
    
    def __init__(self):
        super(Discriminator, self).__init__()
        n_features = 12288
        n_out = 1
        
        self.hidden0 = nn.Sequential(
            nn.Conv2d(in_channels = 3,out_channels = 64, kernel_size = 4, stride = 2,padding = 1,bias=False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(),
            nn.Dropout(0.3)


        )
        self.hidden1 = nn.Sequential(            
            nn.Conv2d(in_channels = 64,out_channels = 128, kernel_size = 4, stride = 2,padding = 1,bias=False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(),
            nn.Dropout(0.3)
        )
        self.hidden2 = nn.Sequential(
            nn.Conv2d(in_channels = 128,out_channels = 256, kernel_size = 4, stride = 2,padding = 1,bias=False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(),
            nn.Dropout(0.3)

        )
        self.hidden3 = nn.Sequential(
            nn.Conv2d(in_channels = 256,out_channels = 512, kernel_size = 4, stride = 2,padding = 1,bias=False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(),
            nn.Dropout(0.3)
        )
        self.out = nn.Sequential(
            nn.Conv2d(in_channels = 512, out_channels = 1, kernel_size = 4, stride = 2,padding = 1,bias=False),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.hidden0(x)
        x = self.hidden1(x)
        x = self.hidden2(x)
        x = self.hidden3(x)
        x = self.hidden4(x)

        return x
    
discriminator = Discriminator()
```

On the other hand the Generator will begin with 100 images and will try to create 12288 images for the discriminator to discriminate.

```python
class Generator(torch.nn.Module):
   
    def __init__(self):
        super(Generator, self).__init__()
        n_features = 100
        n_out = 12288
        loss_func = nn.BCELoss()

        self.hidden0 = nn.Sequential(
            nn.ConvTranspose2d(100,512,kernel_size=4, stride=1,padding=0,bias=False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(),
            nn.Dropout(0.3)


        )
        self.hidden1 = nn.Sequential(            
            nn.ConvTranspose2d(512,256,kernel_size=4, stride=1,padding=0,bias=False),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(),
            nn.Dropout(0.3)
        )
        self.hidden2 = nn.Sequential(
            nn.ConvTranspose2d(256,128,kernel_size=4, stride=1,padding=0,bias=False),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(),
            nn.Dropout(0.3)

        )
        self.hidden3 = nn.Sequential(
            nn.ConvTranspose2d(128,64,kernel_size=4, stride=1,padding=0,bias=False),
            nn.BatchNorm2d(64),
            nn.LeakyReLU(),
            nn.Dropout(0.3)
        )
        self.out = nn.Sequential(
            nn.ConvTranspose2d(64,3,kernel_size=4, stride=1,padding=0,bias=False),
            nn.Tanh()
        )

    def forward(self, x):
        x = self.hidden0(x)
        x = self.hidden1(x)
        x = self.hidden2(x)
        x = self.hidden3(x)
        x = self.out(x)
        return x
generator = Generator()
```

We will be initializing the discriminator and generator weights randomly in each layer based on the type of the layer.

```python
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv2d') != -1:
        m.weight.data.normal_(0.0, 0.02)
    elif classname.find('BatchNorm2d') != -1:
        m.weight.data.normal_(1.0,0.02)
        m.bias.data.fill_(0)

discriminator = discriminator.apply(weights_init)
generator = generator.apply(weights_init)
```
